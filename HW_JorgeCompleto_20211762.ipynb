````python
def statistics(X):
    mean = sum(X)/len(X) # compute the mean
    variance = sum([((i - mean) ** 2) for i in X]) / len(X) # compute the variance using the mean
    std = variance ** 0.5 # compute the standard deviation using the variance
    return (mean, std)

statistics(X)
````

## K-Means from Scratch!

K-Means is one of the most popular "clustering" algorithms, which you perhaps are already familiar from Data Mining or Machine Learning curricular units. K-means estimates the location of $k$ centroids that represent the typical value in each cluster. In that sense, an observation is considered to be part of a particular cluster if it is closer to that cluster's centroid than of any other centroid.

The K-means algorithm is iterative and can be summarized as follows:
1. Start by choosing value K (number of clusters) that defines the number of clusters;
2. Initialize the locations of the centroids, which we can do by randomly select K points from your dataset as initial locations of the centroids;
3. Calculate distance of all other points to each of the K centroids;
4. Associate each point to the cluster of the closest centroid;
5. Update the centroid position, by computing the average coordinates of all points assigned to each cluster;
6. Evaluate the average change in the centroids positions, as a measure of convergence (we assume algorithm converged to the solution when the positions of the centroids don't change more than a given tolerance threshold);
7. Repeat steps 3-6 until either the centroids no longer move more than a tolerance threshold or until you repeated these steps at least for specified number of iterations (niter)

The algorithm is thus parametized by the number of clusters (K), the maximum number of iterations (niter), and the tolerance threshold (tol).


__Important keywords:__
- **points**: data point coordinates `(e.g.: [[0, 0.9], [0.3, 0.5], [0.8, 0.75], [0.1, 0.1]])`
- **centroids**: center of each cluster `(e.g.: [[0.5, 0.5], [0.8, 0.9]])`
- **clusters**: cluster index of each point `(e.g.: [0, 0, 1, 0])`

````python
def get_centroids(points, clusters, k):
    centroids = []
    for i in range(k): # iterate over the number of clusters k 
        aux = [] # auxiliar list 
        for x in range(len(points)): 
            if clusters[x] == i:
                aux.append(points[x]) 
        centroids.append([statistics.mean(n) for n in zip(*aux)]) # compute the mean for each value of the aggregated auxiliar list
    return centroids
````

_NOTE!: A common isue with K-Means is that it can happen that empty clusters are created, e.g. If there are 3 clusters, and all points are either closer to cluster 0 or cluster 1, your code may raise errors when calculating statistics such as cluster 2 centroids or errors.
````python
#Test variables, do not change the values
points = [[0.8922063, 0.26672425],[0.34475611, 0.35976697],[0.33253499, 0.18923898],[0.66872466, 0.46248986],[0.72823733, 0.10537784],[0.40903598, 0.70639412],[0.79926596, 0.90095583],[0.67886544, 0.84573289],[0.3641813, 0.64296743],[0.07461196, 0.74290527]]
k = 2
n_iter = 100
tol = 0.0005
````

````python
def kmeans(points, k, n_iter, tol):
    """
    How kmeans works
    # 1. Set initial centroids/seeds
    # 2. Iterate the number of iterations and, in each one: (until stopping criteria)
    #   2.1. Get the list of clusters (each point will belong to a cluster)
    #   2.2. Compute the centroids of your obtained clusters
    #   2.3. Compute the error between your old centroids and new centroids
    #   2.4. If error is smaller than your tollerance, stop the algorithm
    # 3. Return (clusters, centroids) where clusters = list of each point assigned cluster, and 

    Parameters
    ------------
    points: list of points
    k: number of clusters
    n_iter: number of iterations
    tol: error tolerance threshold, below which your algorithm should stop

    Returns
    ------------
    Tuple (clusters, centroids)

    """
    oldcentroids = get_seeds(points, k) # compute the first set of centroids
    for i in range(n_iter):
        clusters = get_clusters(points, oldcentroids)
        centroids = get_centroids(points, clusters, k)
        error = get_error(oldcentroids, centroids)
        if error < tol:
            break
        oldcentroids = centroids # end of iterations after the error is lower than the tolerable error

    return clusters, centroids
````

````python
def get_inertia(points, clusters, centroids):
    sse = 0
    n = len(clusters) # define the number of clusters
    for i in range(len(points)):
        position = clusters[i]
        sse += distance(points[i], centroids[position]) ** 2 # use the distance function from step 3
    return n ,sse

get_inertia(points = points, clusters = clusters, centroids = centroids)
````

#### Why is inertial useful? (You can just run this, it's independent from the exercises)

It can be used to measure the quality of your clustering. You can determine the optimal number of clusters with a common technique called `The Elbow Plot`, that visually represents the relationship between the number of clusters and the total error.<br>

The more clusters you have, the less error is expected. You want to try to find the optimal threshold where adding more clusters doesn't result in a significant decrease in the total error. This optimal point usually resembles an elbow. 

[Wikipedia](https://en.wikipedia.org/wiki/Elbow_method_(clustering))

````python
# !!! Example - Not graded. Just run.

inertia_example = pd.DataFrame({
    "k": range(1, 11),
    "inertia": [500, 350, 280, 230, 155, 145, 130, 125, 122, 118]
})

plt.figure(figsize=(8, 5))
plt.plot(inertia_example["k"], inertia_example["inertia"], linewidth=3)
plt.xlabel("Clusters")
plt.ylabel("SSE (error)")
plt.show()
````

In this example, notice how the error gets smaller and smaller, as the number of clusters increases, until a point where adding clusters is not worth it.
